````python
import yaml
from IPython.display import display, clear_output
import ipywidgets as widgets
import os

CONFIG_NAME = ""
yaml_directory_path = '/content/FollowYourPose/configs' # TO CHANGE
conf_file = [file for file in os.listdir(yaml_directory_path) if file.endswith('.yaml')]

pretrained_directory_path = '/content/FollowYourPose/checkpoints' # TO CHANGE - depending on where ./checkpoints/stable-diffusion-v1-4 is
# Get a list of all files in the pretrained_directory_path
pretrained_files = [file for file in os.listdir(pretrained_directory_path)]

POSE_SAMPLE_YAML = "/content/FollowYourPose/configs/pose_sample.yaml"
POSE_TRAIN_YAML = "/content/FollowYourPose/configs/pose_train.yaml"

widgets_dict = {}
config = None

# Define widgets for interactive modification
widget_width = "500px"

# Create a dropdown widget
video_dropdown = widgets.Dropdown(
    options=[None] + conf_file,
    description='Select file:'
)

def on_dropdown_change(change):
    global CONFIG_NAME, widgets_dict, config
    # CONFIG_NAME = change['new']
    CONFIG_NAME = os.path.join(yaml_directory_path, change['new'])
    # print(CONFIG_NAME)
    clear_output(wait=True)

    # Load the configuration file and define the 'config' variable
    if CONFIG_NAME == POSE_SAMPLE_YAML:
        print("You have choosen:", change['new'])
        loaded_config_name = POSE_SAMPLE_YAML
        config = yaml.safe_load(open(POSE_SAMPLE_YAML, "r"))
        # Make sure the initial pretrained_model_path is in the list
        initial_pretrained_model_path = config["pretrained_model_path"]
        if initial_pretrained_model_path not in pretrained_files:
            config["pretrained_model_path"] = None  # Set it to None if not in the list

        widget_configs = {
            "pretrained_model_path": {"type": "Dropdown", "value": config["pretrained_model_path"], "description": "Pretrained Model Path:"},
            "output_dir": {"type": "Text", "value": config["output_dir"], "description": "Output Directory:"},

            "validation_data.prompts": {"type": "Textarea", "value": "\n".join(config["validation_data"]["prompts"]), "description": "Prompts:", "layout": {'height': '100px'}},
            "validation_data.video_length": {"type": "IntSlider", "value": config["validation_data"]["video_length"], "min": 0, "max": 32, "step": 1, "description": "Video Length:"},
            "validation_data.width": {"type": "IntSlider", "value": config["validation_data"]["width"], "min": 0, "max": 1024, "step": 1, "description": "Width:"},
            "validation_data.height": {"type": "IntSlider", "value": config["validation_data"]["height"], "min": 0, "max": 1024, "step": 1, "description": "Height:"},
            "validation_data.num_inference_steps": {"type": "IntSlider", "value": config["validation_data"]["num_inference_steps"], "min": 0, "max": 100, "step": 1, "description": "Num Inference Steps:"},
            "validation_data.guidance_scale": {"type": "FloatSlider", "value": config["validation_data"]["guidance_scale"], "min": 0, "max": 20, "step": 0.1, "description": "Guidance Scale:"},
            "validation_data.use_inv_latent": {"type": "ToggleButtons", "options": ["True", "False"], "value": str(config["validation_data"]["use_inv_latent"]), "description": "Use Inverse Latent:", "layout": {'width': 'auto'}},
            "validation_data.num_inv_steps": {"type": "IntSlider", "value": config["validation_data"]["num_inv_steps"], "min": 0, "max": 100, "step": 1, "description": "Num Inverse Steps:"},
            "validation_data.dataset_set": {"type": "Text", "value": config["validation_data"]["dataset_set"], "description": "Dataset Set:"},

            "train_batch_size": {"type": "IntSlider", "value": config["train_batch_size"], "min": 1, "max": 10, "step": 1, "description": "Train Batch Size:"},
            "validation_steps": {"type": "IntSlider", "value": config["validation_steps"], "min": 0, "max": 500, "step": 1, "description": "Validation Steps:"},
            "resume_from_checkpoint": {"type": "Text", "value": config["resume_from_checkpoint"], "description": "Resume from Checkpoint:"},
            "seed": {"type": "IntSlider", "value": config["seed"], "min": 0, "max": 100, "step": 1, "description": "Seed:"},
            "mixed_precision": {"type": "Text", "value": config["mixed_precision"], "description": "Mixed Precision:"},
            "gradient_checkpointing": {"type": "ToggleButtons", "options": ["True", "False"], "value": str(config["gradient_checkpointing"]), "description": "Gradient Checkpointing:", "layout": {'width': 'auto'}},
            "enable_xformers_memory_efficient_attention": {"type": "ToggleButtons", "options": ["True", "False"], "value": str(config["enable_xformers_memory_efficient_attention"]), "description": "Enable Xformers Memory Efficient Attention:", "layout": {'width': 'auto'}},

       }
    elif CONFIG_NAME == POSE_TRAIN_YAML:
        print("You have choosen:", change['new'])
        loaded_config_name = POSE_TRAIN_YAML
        config = yaml.safe_load(open(POSE_TRAIN_YAML, "r"))
        # Make sure the initial pretrained_model_path is in the list
        initial_pretrained_model_path = config["pretrained_model_path"]
        if initial_pretrained_model_path not in pretrained_files:
            config["pretrained_model_path"] = None  # Set it to None if not in the list

        widget_configs = {
            "pretrained_model_path": {"type": "Dropdown", "value": config["pretrained_model_path"], "description": "Pretrained Model Path:"},
            "output_dir": {"type": "Text", "value": config["output_dir"], "description": "Output Directory:"},

            "train_data.n_sample_frames" :{"type": "IntSlider", "value": config["train_data"]["n_sample_frames"], "min": 0, "max": 20, "step": 1, "description": "N Sample Frames:"},
            "train_data.width" :{"type": "IntSlider", "value": config["train_data"]["width"], "min": 0, "max": 1024, "step": 1, "description": "Width:"},
            "train_data.height" :{"type": "IntSlider", "value": config["train_data"]["height"], "min": 0, "max": 1024, "step": 1, "description": "Height:"},
            "train_data.sample_start_idx" :{"type": "IntSlider", "value": config["train_data"]["sample_start_idx"], "min": 0, "max": 10, "step": 1, "description": "Sample Start Idx:"},
            "train_data.sample_frame_rate" :{"type": "IntSlider", "value": config["train_data"]["sample_frame_rate"], "min": 1, "max": 10, "step": 1, "description": "Sample Frame Rate:"},
            "train_data.dataset_set" : {"type": "Text", "value": config["train_data"]["dataset_set"], "description": "Dataset Set:"},

            "validation_data.prompts": {"type": "Textarea", "value": "\n".join(config["validation_data"]["prompts"]), "description": "Prompts:", "layout": {'height': '100px'}},
            "validation_data.video_length": {"type": "IntSlider", "value": config["validation_data"]["video_length"], "min": 0, "max": 32, "step": 1, "description": "Video Length:"},
            "validation_data.width": {"type": "IntSlider", "value": config["validation_data"]["width"], "min": 0, "max": 1024, "step": 1, "description": "Width:"},
            "validation_data.height": {"type": "IntSlider", "value": config["validation_data"]["height"], "min": 0, "max": 1024, "step": 1, "description": "Height:"},
            "validation_data.num_inference_steps": {"type": "IntSlider", "value": config["validation_data"]["num_inference_steps"], "min": 0, "max": 100, "step": 1, "description": "Num Inference Steps:"},
            "validation_data.guidance_scale": {"type": "FloatSlider", "value": config["validation_data"]["guidance_scale"], "min": 0, "max": 20, "step": 0.1, "description": "Guidance Scale:"},
            "validation_data.use_inv_latent": {"type": "ToggleButtons", "options": ["True", "False"], "value": str(config["validation_data"]["use_inv_latent"]), "description": "Use Inverse Latent:", "layout": {'width': 'auto'}},
            "validation_data.num_inv_steps": {"type": "IntSlider", "value": config["validation_data"]["num_inv_steps"], "min": 0, "max": 100, "step": 1, "description": "Num Inverse Steps:"},
            "validation_data.dataset_set": {"type": "Text", "value": config["validation_data"]["dataset_set"], "description": "Dataset Set:"},

            "learning_rate": {"type": "Text", "value": config["learning_rate"], "description": "Learning Rate:"},
            "train_batch_size": {"type": "IntSlider", "value": config["train_batch_size"], "min": 1, "max": 10, "step": 1, "description": "Train Batch Size:"},
            "max_train_steps": {"type": "IntSlider", "value": config["max_train_steps"], "min": 0, "max": 1000, "step": 1, "description": "Max Train Steps:"},
            "checkpointing_steps": {"type": "IntSlider", "value": config["checkpointing_steps"], "min": 0, "max": 1000, "step": 1, "description": "Checkpointing Steps:"},
            "validation_steps": {"type": "IntSlider", "value": config["validation_steps"], "min": 0, "max": 500, "step": 1, "description": "Validation Steps:"},
            "trainable_modules": {"type": "Textarea", "value": "\n".join(config["trainable_modules"]), "description": "Trainable Modules:", "layout": {'height': '100px'}},
            "skeleton_path": {"type": "Text", "value": config["skeleton_path"], "description": "Skeleton Path:"},

            "seed": {"type": "IntSlider", "value": config["seed"], "min": 0, "max": 100, "step": 1, "description": "Seed:"},
            "mixed_precision": {"type": "Text", "value": config["mixed_precision"], "description": "Mixed Precision:"},
            "use_8bit_adam": {"type": "ToggleButtons", "options": ["True", "False"], "value": str(config["use_8bit_adam"]), "description": "Use 8bit Adam:", "layout": {'width': 'auto'}},
            "gradient_checkpointing": {"type": "ToggleButtons", "options": ["True", "False"], "value": str(config["gradient_checkpointing"]), "description": "Gradient Checkpointing:", "layout": {'width': 'auto'}},
            "enable_xformers_memory_efficient_attention": {"type": "ToggleButtons", "options": ["True", "False"], "value": str(config["enable_xformers_memory_efficient_attention"]), "description": "Enable Xformers Memory Efficient Attention:", "layout": {'width': 'auto'}},

        }
    else:
        # print(CONFIG_NAME, POSE_SAMPLE_YAML, POSE_TRAIN_YAML)
        print("You have choosen:",CONFIG_NAME)

    # Display the interactive widgets
    for key, value in widget_configs.items():
        widgets_dict[key] = create_widget(key, value["value"], value["type"], widget_configs=widget_configs)

    display(*widgets_dict.values(), save_button)

# Attach the event handler function
video_dropdown.observe(on_dropdown_change, names='value')


def create_widget(config_key, config_value, widget_type, widget_configs=None, **kwargs):
    widget_style = kwargs.pop('style', {'description_width': 'initial'})

    if widget_type == 'Text':
        return widgets.Text(value=config_value, description=widget_configs[config_key]["description"], style=widget_style, layout={'width': widget_width}, **kwargs)

    elif widget_type == 'Textarea':
        return widgets.Textarea(value=config_value, layout={'height': '100px', 'width': widget_width}, description=widget_configs[config_key]["description"], style=widget_style, **kwargs)

    elif widget_type == 'ToggleButtons':
        options = ["True", "False"]
        widget_value = str(config_value).capitalize()
        return widgets.ToggleButtons(options=options, value=widget_value, description=widget_configs[config_key]["description"], style=widget_style, **kwargs)

    elif widget_type == 'IntSlider':
        max_value = widget_configs[config_key].get("max", 100)
        return widgets.IntSlider(min=kwargs.get('min', 0), max=max_value,
                                step=kwargs.get('step', 1), value=config_value, description=widget_configs[config_key]["description"],
                                style=widget_style, layout={'width': widget_width}, **kwargs)

    elif widget_type == 'FloatSlider':
        max_value = widget_configs[config_key].get("max", 100.0)
        return widgets.FloatSlider(min=kwargs.get('min', 0), max=max_value,
                                  step=kwargs.get('step', 0.1), value=config_value, description=widget_configs[config_key]["description"],
                                  style=widget_style, layout={'width': widget_width}, **kwargs)

    elif widget_type == 'Dropdown':
        return widgets.Dropdown(options=pretrained_files, value=config_value, description=widget_configs[config_key]["description"], style=widget_style, layout={'width': widget_width}, **kwargs)

    else:
        return widgets.Text(value=f'Widget type not recognized: {widget_type}', description=widget_configs[config_key]["description"], style=widget_style)





# Define a button for saving changes
save_button = widgets.Button(description="Save Changes")

# Define a function to handle save button click event
def save_changes(button):
    global CONFIG_NAME, config

    # Check if CONFIG_NAME is empty
    if not CONFIG_NAME:
        print("Please select a configuration file first.")
        return

    error_message = None  # Initialize error message variable

    for key, widget in widgets_dict.items():
        keys = key.split('.')
        current_config = config

        # Traverse the nested dictionary to get to the right level
        for k in keys[:-1]:
            current_config = current_config[k]

        # Handle different widget types
        if isinstance(widget, widgets.Textarea):
            current_config[keys[-1]] = [prompt.strip() for prompt in widget.value.split('\n')]

        elif isinstance(widget, (widgets.IntSlider, widgets.FloatSlider)):
            current_config[keys[-1]] = widget.value

        elif isinstance(widget, widgets.ToggleButtons):
            current_config[keys[-1]] = widget.value

        elif isinstance(widget, widgets.Dropdown):
            # Save the full path for the Dropdown widget
            # current_config[keys[-1]] = os.path.join(pretrained_directory_path, widget.value)
            if widget.value is None:
                error_message = "Error: Pretrained Model Path cannot be None."
            else:
                current_config[keys[-1]] = os.path.relpath(os.path.join(pretrained_directory_path, widget.value), start='/content/FollowYourPose')

        else:
            # For other widget types, set the value directly
            current_config[keys[-1]] = widget.value

    # Save the updated configuration back to the YAML file
    if error_message:
        print(error_message)
    else:
        # Save the updated configuration back to the YAML file
        with open(CONFIG_NAME, "w") as config_file:
            yaml.dump(config, config_file, default_flow_style=False)

        print("Changes saved successfully!")


# Attach the save function to the button click event
save_button.on_click(save_changes)

display(video_dropdown)

````
