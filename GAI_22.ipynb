```python
#@title Extract animation with inferencer
from IPython.display import HTML, display, clear_output
import ipywidgets as widgets
import os
from base64 import b64encode
import time

# Function to display a video using base64 encoding
def show_video(video_path, video_width=600):
    try:
        video_file = open(video_path, "rb").read()
        video_url = f"data:video/mp4;base64,{b64encode(video_file).decode()}"
        return HTML(f'<video width="{video_width}" controls><source src="{video_url}"></video>')
    except Exception as e:
        return HTML(f'Error: {str(e)}')

# Define the directory where the Charades videos are located
video_dir = "/content/Trimed_video_input/video"

# List all the video files in the directory
video_files = [f for f in os.listdir(video_dir) if f.endswith(".mp4")]

# Check if there are any video files in the directory
if len(video_files) > 0:
    # Create a dropdown widget for video selection
    video_selection = widgets.Dropdown(
        options=video_files,
        description='Select a video:',
        layout=widgets.Layout(width='400px')
    )

    # Create a dropdown widget for method selection
    method_dropdown = widgets.Dropdown(
        options=["2D Video Test", "3D Video Test", "2D Whole Body Video Test"],
        description="Select a method:"
    )

    # Create a button widget
    run_button = widgets.Button(description="Run Selected Method")

    # Define event handlers for the button click
    def on_button_click(b):
        selected_video = os.path.join(video_dir, video_selection.value)
        video_filename = os.path.splitext(os.path.basename(selected_video))[0]  # Remove ".mp4" extension
        selected_method = method_dropdown.value
        clear_output()
        print(f"Running {selected_method} on selected video: {selected_video}")

        # Run the selected method on the selected video
        vis_out_dir = os.path.join("vis_results", video_filename)  # Construct vis-out-dir
        if selected_method == "2D Video Test":
            # Replace the --input path with the selected video path
            !python demo/inferencer_demo.py \
                "{selected_video}" \
                --pose2d human --vis-out-dir "{vis_out_dir}" --pred-out-dir "{vis_out_dir}" --black-background

        elif selected_method == "3D Video Test":
            # Replace the --input path with the selected video path
            !python demo/inferencer_demo.py \
                "{selected_video}" \
                --pose3d human3d --vis-out-dir "{vis_out_dir}" --pred-out-dir "{vis_out_dir}" --black-background

        elif selected_method == "2D Whole Body Video Test":
            # Replace the --input path with the selected video path
            !python demo/inferencer_demo.py \
                "{selected_video}" \
                --pose2d wholebody --vis-out-dir "{vis_out_dir}" --pred-out-dir "{vis_out_dir}" --black-background

    # Attach the event handler to the button
    run_button.on_click(on_button_click)

    # Display the video selection dropdown and method selection dropdown
    display(video_selection)
    display(method_dropdown)

    # Display the run button
    display(run_button)
else:
    print("No video files found in the directory.")
```

```python
#GAI-9
import os
import psutil
import subprocess
import time
from ipywidgets import widgets, Output, interactive, HBox
from IPython.display import display, Video, clear_output
import sys

# Function to print CPU and memory usage
def print_system_resources():
    cpu_percent = psutil.cpu_percent()
    memory_info = psutil.virtual_memory()

    print(f"CPU Usage: {cpu_percent}%")
    print(f"Memory Usage: {memory_info.percent}%")

# Specify the path to the directory containing the video files
directory_path = '/content/Trimed_video_input/video/'

# Get a list of all files in the directory
video_files = [file for file in os.listdir(directory_path) if file.endswith('.mp4')]

if(len(video_files)<=0):
  print("No video files found in the directory.")
  sys.exit(0)

# Create a dropdown widget
video_dropdown = widgets.Dropdown(
    options=video_files,
    description='Select Video:'
)

# Create a text box widget
text_input = widgets.Text(
    description='Enter Text:',
    style={'description_width': 'initial'}  # Set the width to 'initial'
)
# Create a button widget
button = widgets.Button(
    description='Submit Text'
)
# Arrange the text_input and button widgets horizontally
text_input_and_button = HBox([text_input, button])

# Create an Output widget to display the video player
video_output = Output()
# Create an Output widget for displaying the "Selected Video" message
selected_video_output = Output()
# Create an Output widget for displaying the "You have entered" message
selected_textinput_output = Output()

# Set a fixed size for the video player
video_width = 320
video_height = 240

selected_video = ""

# Function to handle dropdown value changes
def on_dropdown_change(change):
    selected_video = change['new']

    # Clear the previous output in the dedicated output area for the "Selected Video" message
    with selected_video_output:
        clear_output(wait=True)
        # Display the "Selected Video" message
        print(f"Selected Video: {selected_video}")

    # start_time = time.time()

    video_path = os.path.join(directory_path, selected_video)
    # Resize the video to a smaller dimension before displaying
    resized_video_content = subprocess.check_output(['ffmpeg', '-i', video_path, '-vf', 'scale=320:240', '-vcodec', 'rawvideo', '-f', 'rawvideo', '-'], stderr=subprocess.STDOUT)

    # To run if video is taking too long to load
    # Print system resources after loading the video
    # print_system_resources()
    # elapsed_time = time.time() - start_time
    # print(f"Time to load video: {elapsed_time:.2f} seconds")

    # Clear the previous output in the dedicated output area for the video player
    with video_output:
        clear_output(wait=True)
        # Display the video player
        display(Video(video_path, embed=True, width=video_width, height=video_height))


# Function to handle button click
def on_button_click(b):
    entered_text = text_input.value
    with selected_textinput_output:
      clear_output(wait=True)
      print(f"You have entered: {entered_text}")
      selected_video = os.path.join(directory_path, video_dropdown.value)
      video_filename = os.path.splitext(os.path.basename(selected_video))[0]  # Remove ".mp4" extension
      selected_method = method_dropdown.value
      clear_output()
      print(f"Running {selected_method} on selected video: {selected_video}")

      # Run the selected method on the selected video
      vis_out_dir = os.path.join("vis_results", video_filename)  # Construct vis-out-dir
      if selected_method == "2D Video Test":
          # Replace the --input path with the selected video path
          !python demo/inferencer_demo.py \
              "{selected_video}" \
              --pose2d human --vis-out-dir "{vis_out_dir}" --pred-out-dir "{vis_out_dir}" --black-background

      elif selected_method == "3D Video Test":
          # Replace the --input path with the selected video path
          !python demo/inferencer_demo.py \
              "{selected_video}" \
              --pose3d human3d --vis-out-dir "{vis_out_dir}" --pred-out-dir "{vis_out_dir}" --black-background

      elif selected_method == "2D Whole Body Video Test":
          # Replace the --input path with the selected video path
          !python demo/inferencer_demo.py \
              "{selected_video}" \
              --pose2d wholebody --vis-out-dir "{vis_out_dir}" --pred-out-dir "{vis_out_dir}" --black-background

# Create a dropdown widget for method selection
method_dropdown = widgets.Dropdown(
    options=["2D Video Test", "3D Video Test", "2D Whole Body Video Test"],
    description="Select a method:"
)


# Attach the function to the dropdown's value change event
print(video_dropdown.observe(on_dropdown_change, names='value'))
button.on_click(on_button_click)
# print_system_resources()

# Display the dropdown widget and the dedicated output area
display(video_dropdown)
display(method_dropdown)
display(video_output)
display(selected_video_output)

display(text_input_and_button)
display(selected_textinput_output)  # Display the output area for entered text

# Optionally, display an initial video player (comment this part if not needed)
# if video_files:
#     initial_video_path = os.path.join(directory_path, video_files[0])
#     with video_output:
#         display(Video(initial_video_path, embed=True, width=video_width, height=video_height))
```

```python
import os
import ipywidgets as widgets
from IPython.display import display

def get_image(f_path):
    image_type = ['gif']
    if any(x in image_type for x in f_path.split('.')):
        file = os.path.join(folder, f_path)
        image = open(file, 'rb').read()
        return image

folder = '/content/FollowYourPose/gif_results/' #Need to set actual FollowYourPose output here
files = os.listdir(folder)
images = [get_image(x) for x in files]
valid_images = [img for img in images if img is not None]
file_names = [file_name for img, file_name in zip(images, files) if img is not None]
children = [widgets.Image(value=img, width='auto', height='auto', layout=widgets.Layout(margin='auto')) for img in valid_images]

# Create a list of VBox to display image and file name vertically
content_widgets = [widgets.VBox([child, widgets.HTML(f"<div style='font-size: 24px; text-align: center;'>{file_name}</div>")]) for child, file_name in zip(children, file_names)]

box_layout = widgets.Layout(
    display='flex',
    flex_flow='column',
    align_items='stretch',
    border='solid',
    width='100%')

# Create the widget
tab = widgets.Tab()
tab.children = content_widgets
tab.layout = box_layout

# Set the labels as file names
for i, file_name in enumerate(file_names):
    tab.set_title(i, file_name)

display(tab)
```
